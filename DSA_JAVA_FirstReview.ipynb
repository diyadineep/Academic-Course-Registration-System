## Introduction
The Academic Course Registration System is designed to manage the essential aspects of course registration in an academic institution. It provides a robust platform for organizing courses, managing prerequisites, and facilitating student enrollments. The system ensures efficient handling of data and streamlines the process of searching, validating, and registering for courses.

## Purpose of the System
The purpose of this project is to create a reliable and efficient system that addresses the key challenges of academic course management:

1. Organized Course Catalogue: Maintain a structured collection of courses for easy management and access.
2. Prerequisite Validation: Ensure that students meet the required prerequisites before registering for advanced courses.
3. Efficient Registration Process: Allow students to quickly search for and register in courses without conflicts or errors.

## Objectives
The system is built around three primary goals:

**1. Binary Search Tree (BST) for Course Catalogue:**
* Use a BST to organize courses by their unique course codes.
* The BST enables efficient operations such as adding, removing, and searching for courses based on unique course codes.

**2. Management of Prerequisites and Student Enrollments:**
* Each course can have a defined set of prerequisites to guide academic progression.

* The system tracks student enrollments, preventing duplicate registrations and errors.

**3. Efficient Course Search and Registration:**
* Provides quick and flexible search options (e.g., by course code or name).
* Simplify the registration process for students by integrating prerequisite checks.

```python
  class Student:
        def __init__(self, student_id, student_name, email):
           self.student_id = student_id
           self.student_name = student_name
           self.email = email
           self.registered_courses = []
  

* **Registration** : Stores student course registration details with a timestamp.
```python
 class Registration:
      def __init__(self, registration_student, registration_course, registration_time):
        self.registration_student = registration_student
        self.registration_course = registration_course
        self.registration_time = registration_time

* **CourseRegistration** : A singleton class that manages courses, students, and registrations.
```python
class CourseRegistrationSystem:
    instance = None
    
    def _init_(self):
        self.courses: Dict[str, Course] = {}
        self.students: Dict[int, Student] = {}
        self.registrations: List[Registration] = []

### Attributes
Attributes are variables that store object-specific data.
* **Course Attributes** : course_id, course_name, max_capacity, enrolled_students.
* **Student Attributes** : student_id, student_name, email, registered_courses.
* **Registration Attributes** : registration_student, registration_course, registration_time.
* **CourseRegistrationSystem Attributes** : _instance, courses, students, registrations.



### Data Structures Used
#### **Binary Search Tree**: for course catalogue
Why BST:
* It allows fast searching, insertion, and deletion of courses.
* Searches take O(log n) time, unlike arrays which take O(n) in the worst case.
* Courses are stored in alphabetic order, making it easy to find the next or previous course
* New courses can be added or removed efficiently.

How BST is used:
* Adding Courses: New courses are placed in the tree based on their course codes.
* Searching Courses: Students can quickly find a course in O(log n) time.
* Removing Courses: Courses can be deleted while keeping the tree balanced.
* Listing Courses: An inorder traversal shows all courses in sorted order.

  
Other Data Structures :
**AVL Tree** - 
* AVL tree keeps courses sorted, allowing quick searches using course codes or names in O(log n) time. It ensures fast insertion and deletion by balancing itself automatically. This prevents the tree from becoming unbalanced and maintains efficiency while keeping courses in alphabetical order.
  

#### **Graphs**: for managing course prerequisites and student enrolments
Why Graphs:
* It easily represents course dependencies. 
* Ensures no circular dependencies in prerequisites.
* Helps determine the order in which courses can be taken.
* Using an adjacency list, checking prerequisites takes O(1) – O(n) time.

How is it used:
* Each course is a node in the graph.
* Directed edges represent prerequisites.
* When a student wants to enroll in a course, the system checks the graph to ensure all prerequisites are met.


#### **Binary Search Tree**: for supporting efficient course search & registration
why BST:
* BST provides O(log n) search 
* BST maintains courses in alphabetical or ID-based order for listing.

Other Data structures used:
**Hash Table**
* Hash Tables provide fast O(1) time search.
* Hash Tables allow instant course enrollment lookups.
* But it does not give a sorted order.


#### **Queue**: for handling waitlists for full courses
Why Queue:
* FIFO Enrollment – First-come-first-serve approach ensures fairness.
* Queue operations are O(1) therefore provide efficient processing.

How is it used:
* If a student tries to register for a full course, they are placed in a Queue (FIFO order).
* When a seat opens, the first student in the queue is enrolled automatically.


Other Data Structures used:
**Min-Heap (Priority Queue)**:
* Best for priority-based waitlists. Students with higher priority get enrolled first.
